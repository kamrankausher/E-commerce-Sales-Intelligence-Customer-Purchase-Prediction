## 1. Load All Datasets

The Olist dataset consists of **8 relational tables** covering the entire e-commerce lifecycle:

| Table | Description |
|-------|-------------|
| `customers` | Customer demographics (ID, city, state) |
| `orders` | Order metadata (status, timestamps) |
| `items` | Individual items within each order |
| `payments` | Payment details (type, value, installments) |
| `reviews` | Customer review scores |
| `products` | Product catalog |
| `sellers` | Seller information |
| `category_translation` | Portuguese → English category names |

````
# Load all CSV files
dfs = load_all()
````

## 4. Data Relationships (Entity-Relationship)

```
customers ─── customer_id ───→ orders
                                  │
                    ┌─────────────┼──────────────┐
                    ↓             ↓              ↓
                  items       payments        reviews
                    │
              ┌─────┴─────┐
              ↓           ↓
          products     sellers
              │
              ↓
    category_translation
```

**Key relationships:**
- One customer can place many orders (1:N)
- One order can have many items (1:N)
- One order has one or more payments (1:N)
- One order has zero or one review (1:1)

````
# Verify key relationships
print("Key Relationship Checks:")
print(f"  Unique customers: {dfs['customers']['customer_unique_id'].nunique():,}")
print(f"  Unique orders:    {dfs['orders']['order_id'].nunique():,}")
print(f"  Unique products:  {dfs['products']['product_id'].nunique():,}")
print(f"  Unique sellers:   {dfs['sellers']['seller_id'].nunique():,}")
print(f"  Total items:      {len(dfs['items']):,}")
print(f"  Total payments:   {len(dfs['payments']):,}")
print(f"  Total reviews:    {len(dfs['reviews']):,}")
````

## 5. Initial Observations

After exploring the dataset, here are the key takeaways:

1. **Scale**: ~8,500 customers, ~18,000 orders, 5,000 products, 220 sellers
2. **Order Status**: ~85% delivered, ~5% shipped, ~5% canceled
3. **Reviews**: J-shaped distribution — most are 4-5 stars (positive skew)
4. **Payments**: Credit card dominates (~73%), followed by boleto (~19%)
5. **Geography**: Data covers all 27 Brazilian states with São Paulo dominant
